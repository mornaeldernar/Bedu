````python
fig, axes = plt.subplots(2, 4,figsize=(10, 10), sharex=True, sharey=True);

sns.barplot(x=ct.columns, y=ct.iloc[0], ax=axes[0, 0])
sns.barplot(x=ct.columns, y=ct.iloc[1], ax=axes[0, 1])
sns.barplot(x=ct.columns, y=ct.iloc[2], ax=axes[0, 2])
sns.barplot(x=ct.columns, y=ct.iloc[3], ax=axes[0, 3])
sns.barplot(x=ct.columns, y=ct.iloc[4], ax=axes[1, 0])
sns.barplot(x=ct.columns, y=ct.iloc[5], ax=axes[1, 1])
sns.barplot(x=ct.columns, y=ct.iloc[6], ax=axes[1, 2])
sns.barplot(x=ct.columns, y=ct.iloc[7], ax=axes[1, 3])


axes[0,0].bar_label(axes[0,0].containers[0],fmt='%.1f')
axes[0,1].bar_label(axes[0,1].containers[0],fmt='%.1f')
axes[0,2].bar_label(axes[0,2].containers[0],fmt='%.1f')
axes[0,3].bar_label(axes[0,3].containers[0],fmt='%.1f')
axes[1,0].bar_label(axes[1,0].containers[0],fmt='%.1f')
axes[1,1].bar_label(axes[1,1].containers[0],fmt='%.1f')
axes[1,2].bar_label(axes[1,2].containers[0],fmt='%.1f')
axes[1,3].bar_label(axes[1,3].containers[0],fmt='%.1f')

fig.suptitle('Numero de tipo de propiedades por region', fontsize=15);
````
