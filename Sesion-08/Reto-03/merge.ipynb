No es muy práctico tener las tablas `occupations` y `age_range` por sí solas, ya que la información que contienen está íntimamente relacionada con al información de la tabla `users`.
```python
# Tu código va aquí
#
# ...
# ...
import pandas as pd
```

```python
users = pd.read_csv('https://raw.githubusercontent.com/beduExpert/Procesamiento-de-Datos-con-Python-Santander_2020_2021/master/Datasets/MovieLens/users-raw.csv',index_col=0,)
```

```python
occupations = pd.read_csv('https://raw.githubusercontent.com/beduExpert/Procesamiento-de-Datos-con-Python-Santander_2020_2021/master/Datasets/MovieLens/occupations-raw.csv',index_col=0,)
```

```python
age_range = pd.read_csv('https://raw.githubusercontent.com/beduExpert/Procesamiento-de-Datos-con-Python-Santander_2020_2021/master/Datasets/MovieLens/age_ranges-raw.csv',index_col=0,)
```

```python
users.head(3)
```

```python
occupations.head(3)
```

```python
age_range.head()
```

```python
users_full = pd.merge(users, occupations, left_on='occupation', right_index=True).sort_index()
```

```python
users_full = users_full.rename(columns={'occupation': 'occupation_id', 'description': 'occupation'})
```

```python
users_full.head()
```

```python
users_full = pd.merge(users_full, age_range, left_on='age', right_index=True).sort_index()
```

```python
users_full = users_full.rename(columns={'age': 'age_id', 'range': 'age_range'})
```

```python
users_full
```

```python
users_full.to_csv('users_full.csv')
```
